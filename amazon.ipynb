```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('amazon_prime_titles.csv')

# Show first 10 rows
print(df.head(10))

# Drop missing important values
df = df.dropna(subset=['type','release_year','rating','duration'])

# Count types (Movies vs TV Shows)
type_counts = df['type'].value_counts()

# Plot
plt.figure(figsize=(6,8))
plt.bar(type_counts.index, type_counts.values, color='red')
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Amazon Prime: Movies vs TV Shows')
plt.show()

```
